```python
import numpy as np
import math
import random
n = 7
B = 30
v = np.array([12,9,8,7,5,4,3])
w = np.array([8,6,7,5,4,7,3])
print(max(v/w))
good_i =[]
max_value = 0
for i in np.linspace(0,max(v/w),1000):
    tmp_capa = 0
    tmp_value = 0
    for j in range(n):
        if(v[j] >= i*w[j]):
            tmp_capa+=w[j]
            tmp_value+=v[j]
    if(tmp_capa <= B):
        if(tmp_value > max_value):
            good_i = []
            good_i.append(i)
            max_value = tmp_value
        elif(tmp_value == max_value):
            good_i.append(i)
print(good_i)
print(max_value)
#calculate the gap
a=[]
for i in range(2000):
    lam = random.uniform(0,max(v/w))
    tmp_capa = 0
    tmp_value = 0
    for j in range(n):
        if(v[j] >= lam*w[j]):
            tmp_capa+=w[j]
            tmp_value+=v[j]
    if(tmp_capa <= B):
        a.append(tmp_value)
optigap = (41-np.mean(a))/41
print(np.mean(a))
print(optigap)
```
