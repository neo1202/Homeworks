```python
#載入所需函示庫
from matplotlib import pyplot as plt
%matplotlib inline
# 設定圖形大小; DPI越大圖越大
plt.rcParams["figure.dpi"] = 80
import seaborn as sns
import pandas as pd
import numpy as np
import math
import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.stats.api as sms
import statsmodels.formula.api as smf

def outlier(data_k):
    Q1 = percentile(data_k, 25)
    Q2 = percentile(data_k, 50)
    Q3 = percentile(data_k, 75)
    IQR =Q3 - Q1    #IQR is interquartile range. 
    print("Q1 = ", Q1)
    print("Q2 = ", Q2)
    print("Q3 = ", Q3)
    print("IQR = ", IQR)
    filter = (data_k < Q1 - 1.5 * IQR) | (data_k > Q3 + 1.5 * IQR)
    print("Outliers are listed as follows \n")
    print(data_k.loc[filter])
    outlierArray = data_k.loc[filter]
    
def percentile(data1, p):
    if type(data1) == np.ndarray:        
        alldata = data1.copy()
        data1 = data1.copy()
    else:
        alldata = data1.values.copy()
        data1 = data1.values.copy()
    alldata.sort()
    n = alldata.size
    l = (n + 1) * p / 100 - 1
    f_l = math.floor(l)
    c_l = math.ceil(l)
    percentile_v = alldata[f_l] + (alldata[c_l] - alldata[f_l]) * (l - f_l)
    return percentile_v
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python
df_1= pd.read_excel('Stat_2020_1_Mid.xlsx', sheet_name = "Hourly")
#print(max(df_1['Earnings']))
bin_list = np.linspace(min(df_1['Earnings']), max(df_1['Earnings']), 9) #九個代表8塊
#print(bin_list)
fig, ax = plt.subplots()
counts, bins, patches = plt.hist(df_1['Earnings'], bins=bin_list, density=False, facecolor='g', alpha=0.4)
plt.xlabel('Earnings')
plt.ylabel('Frequency')
plt.title('Histogram of Earnings')
plt.grid(True)
plt.xticks(bin_list)

autolabel(ax.patches)
plt.show()
print("the Histogram is positively skewed with a right tail")
```

```python
print("mean: ", np.mean(df_1['Earnings'])) #平均, 中位數, 眾數
print("median: ", np.median(df_1['Earnings']))
print("mode: ", stats.mode(df_1['Earnings'])[0][0]) 
print("the model class is from 17275 to 19592")
print("because mean>median>mode, the data is slightly positively skewed as the histogram")
```

```python
print(stats.percentileofscore(df_1['Earnings'], 26970, kind = "strict")) #數字找%數
n = percentile(df_1['Earnings'], 95)
print(n)
```

```python
from scipy.stats import moment
m2 = moment(df_1['Earnings'], moment=2)
m3 = moment(df_1['Earnings'], moment=3)
m4 = moment(df_1['Earnings'], moment=4)
g1 = m3 / math.sqrt((m2**3))
print("Coefficients of Skewness = ", g1) # 偏態係數>0 代表slightly skewed to the right
g2 = m4 / math.sqrt((m2**4))
print("Coefficient of Kurtosis = ", g2) # 峰度>3 narrow tall //wide low
#符合在上面證實過的
```

```python
#bell shape and unimodal -> empirical rule
std = np.std(df_1['Earnings'], ddof = 0)
print(df_1['Earnings'].mean() - std)
#68% of Earnings fall in between ()
#empirical : 68, 95, 99.7
#chebyshev : 0,  75, 89
```

## 第二題

```python
df_2= pd.read_excel('Stat_2020_1_Mid.xlsx', sheet_name = "Life_GDP")
fig, ax = plt.subplots()
plt.scatter(df_2['GDP'], df_2['Life Expectancy'], color = 'grey')
plt.title('Scatter plot for GDP vs LE')
plt.ylabel('Life Expectancy')
plt.xlabel('GDP')
#According to the scatter plot, there seem to be a nonlinear relationship between GDP & LE
```

```python
fig, ax = plt.subplots()
df_2['lnGDP'] = 0
for i in df_2.index:
    df_2.loc[i, 'lnGDP'] = math.log(df_2.loc[i, 'GDP'])
print(df_2)
#plt.scatter(np.log(df_2['GDP']), df_2['Life Expectancy'], color = 'grey')
plt.scatter(df_2['lnGDP'], df_2['Life Expectancy'])
plt.title('Scatter plot for lnGDP vs LE')
plt.ylabel('Life Expectancy')
plt.xlabel('lnGDP')
_ = sns.regplot(x='lnGDP', y= 'Life Expectancy', data = df_2, color = 'grey', ci = None)
plt.show()
```

```python
# 共變異數（英語：Covariance），在機率論與統計學中用於衡量兩個隨機變數的聯合變化程度 0代表無關
print(np.cov(df_2['lnGDP'], df_2['Life Expectancy']))
#為12.44, 代表兩者確實正相關
```

```python
print(np.corrcoef(df_2['lnGDP'], df_2['Life Expectancy']))
#coefficient of correlation 為相關係數落在-1~1, cov/ stdx*stdy
#strong positive relationship
```
