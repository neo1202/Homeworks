```python
import gurobipy as gp 
from gurobipy import *
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
p1 = Model("p1")
p1.setParam("NonConvex", 2)
l = p1.addVar(    lb = (111.8/5), vtype = GRB.CONTINUOUS, name = 'L')
x = p1.addVars(6, lb = 0, vtype = GRB.CONTINUOUS, name = 'x')
y = p1.addVars(6, lb = 0, vtype = GRB.CONTINUOUS, name = 'y')

p1.addConstr(x[0], GRB.EQUAL, 0, "x0=0")
p1.addConstr(x[5], GRB.EQUAL, 100, "x5=100")
p1.addConstr(y[0], GRB.EQUAL, 100, "y0=100")
p1.addConstr(y[5], GRB.EQUAL, 150, "y5=150")
p1.setObjective( quicksum(y[i] for i in range(1,5)), GRB.MINIMIZE) 
p1.addConstrs( ( (x[i]<=x[i+1]) for i in range(5) ),"xi<=x(i+1)")
for i in range(5):
    p1.addConstr(((x[i+1]-x[i])**2+(y[i+1]-y[i])**2), GRB.EQUAL, l**2, "every line = L")
p1.addConstrs(     ( (x[i] >=0) for i in range(6))   , "xi >= 0" )
p1.addConstrs(     ( (y[i] >=0) for i in range(6))   , "yi >= 0" )
p1.optimize()
for var in p1.getVars():
    print(var.varName, '= %.2f' %var.x)
print("\n")
print("objective value(max w) = %.2f" %p1.objVal)
```
