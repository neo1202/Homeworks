# 先比兩母體的variance再看平均數是否有差距 有些題目只想知道變異數是否同

```python
### H0 : var1/var2 = 1
### H1 : var1/var2 ≠ 1
#有時候要用onetail記得改
df1 = pd.read_excel('Xr13-34.xlsx')
#display(df1.head())
display(df1.tail())
# 轉換成numpy array
Fixed = df1['Fixed'].values
#print("Fixed = \n", Fixed)
Commission = df1['Commission'].values
#print("Commission = \n", Commission)

df_num = Fixed.shape[0] - 1
df_denom = Commission.shape[0] - 1
print("df_num = ", df_num)
print("df_denom = ", df_denom)
var_ratio = stats.describe(Fixed).variance / stats.describe(Commission).variance
print(f"var_ratio (F value) = {var_ratio:.4f}")
ptmp = stats.f.cdf(var_ratio, df_num, df_denom)
p_value = ptmp * 2
print(f"p_value = {p_value:.4f}")
```

The p-value for equal-variance test is 0.8252, which is larger than the required 0.05 significance level, there is no sufficient evidence to reject the null hypothesis. As a result, we assume that the two populations have equal variance and apply equal-variance t-test on μ1-μ2 = 0 <br>
H0:μ1-μ2>=0 , <br>
H1:μ1-μ2<0
這邊記得解釋μ1,μ2是幹嘛的

```python
# t-test using sms.CompareMeans
cm = sms.CompareMeans(sms.DescrStatsW(Fixed), sms.DescrStatsW(Commission))
# usevarstr, ‘pooled’ or ‘unequal’
cm.summary(usevar='pooled', alpha=0.05)
```

Based on the above result, the one-sided p-value should be 0.038/2 = 0.019, and that is lesser than our significant level. So we have enough evidence at a significance level of 0.05 to reject the null hypothesis in favor of the alternative one and conclude that the commission salesperson outperforms the fixed-salary salesperson.


# 遇到matched pairs作法 用t-test, 畫圖測驗可不可用

```python
df1 = pd.read_excel('Xr13-95.xlsx')
display(df1.tail())
samp1 = df1['This Year'].values
samp2 = df1['Last Year'].values

samp_diff = samp1 - samp2
nobs = samp_diff.shape[0]
df = nobs - 1
print("degree of freedom = ", df)

tmpdesc = stats.describe(samp_diff)
t_value = tmpdesc.mean / (tmpdesc.variance ** 0.5) * (nobs ** 0.5)
print(f"t-value = {t_value:.4f}")
# tmpdesc

# p-values
ptmp = stats.t.cdf(t_value, df)
p_value_left = ptmp
p_value_right = 1 - ptmp
if ptmp > 0.5:
    ptmp = 1 - ptmp
p_value_twoside = ptmp * 2

print(f"p_value (two sided) = {p_value_twoside:.4f}")
print(f"p_value (right tail) = {p_value_right:.4f}")
print(f"p_value (left tail) = {p_value_left:.4f}")
```

```python
#Use a significance level of 0.10
t_alpha = stats.t.ppf(1 - 0.10 / 2, df)
std_xbar = (tmpdesc.variance / nobs) ** 0.5
LCL = tmpdesc.mean - t_alpha * std_xbar
UCL = tmpdesc.mean + t_alpha * std_xbar
print("(a.) Improvement in sales between two years under 90 percent confidence is = [%0.4f, %0.4f] thousands dollar." % (LCL, UCL))
```

H0:μ1-μ2<=0 , <br>
H1:μ1-μ2>0 <br> (記得要說分別是什麼）<br>
(b.) From the result above, we can show that in the right tail test, the p_value is only 0.0001, which is much smaller than our significant level. So we have overwhelming evidence to reject the null hypothesis in favor of the alternative one and conclude that advertising in the Yellow Pages improves sales.<br>

# Nominal Data，一個數值在整個母體中的比例 比較兩個母體

```python
### H0 : p1-p2<=0
### H1 : p1-p2>0
p1 = 47/229
p2 = 25/178
p_pool = (47+25)/(229+178)
sd_p_pool = (p_pool * (1 - p_pool) * 
             (1 / 229 + 1 / 178)) ** 0.5
z_value = (p1 - p2) / sd_p_pool
print(f"p1 = {p1:.4f}")
print(f"p2 = {p2:.4f}")
print(f"sd_p_pool = {sd_p_pool:.4f}")
print(f"z-value = {z_value:.4f}")
p_value = 1 - stats.norm.cdf(z_value)
print(f"p-value = {p_value:.4f}")
```

Because p-value < 0.1, reject H0. There is sufficient evidence to conclude at the 10% significance level, that those who paid the regular price are more likely to buy an extended warranty. <br>
或是底下那樣<br>
H0 : p1-p2<=0
H1 : p1-p2>0

```python
df3 = pd.read_excel('Xr13-156.xlsx')
print(df3['Religion'].value_counts())
print("Tail of dataset:")
display(df3.tail())
No_R = df3['No religion']
No_R = No_R == 1
Y_R = df3['Religion']
Y_R = Y_R == 1

p1 = No_R.mean()
p2 = Y_R.mean()
R12 = np.concatenate([No_R, Y_R])
p_pool = R12.mean()
sd_p_pool = (p_pool * (1 - p_pool) * 
             (1 / No_R.shape[0] + 1 / Y_R.shape[0])) ** 0.5
z_value = (p1 - p2) / sd_p_pool
print(f"p1 = {p1:.4f}")
print(f"p2 = {p2:.4f}")
print(f"sd_p_pool = {sd_p_pool:.4f}")
print(f"z-value = {z_value:.4f}")
p_value = 1 - stats.norm.cdf(z_value)
print(f"p-value = {p_value:.4f}")
print("Since p-value = 0.0702 > 0.05, can't reject the null hypothesis .There is insufficient evidence to infer that at the 5% significance level, that resumes including a reference to religious activity reduces the probability of a call back.")
```

## 一個變因anova
let μi = the average grade in school i<br>
H0: μ1=μ2=μ3=μ4<br>
H1: at least two means differ

```python
df1 = pd.read_excel('Xr14-09.xlsx')
display(df1.tail())
# wide_to_long假設同樣資料在不同因素下的名稱有相同的字頭，因此我們要改欄位名稱。
df1 = df1.rename(columns={'School A': 'sale_A', "School B": "sale_B", "School C": "sale_C", "School D ": "sale_D"})
# 需要一個ID欄位
df1["id"] = df1.index
df2 = pd.wide_to_long(df1, ["sale"], sep="_", suffix='\w+', i="id", j="factor").reset_index()
#print(type(df2))
df2 = df2[~pd.isnull(df2['sale'])]
display(df2)
df2 = df2[['factor', 'sale']]
df2.describe()
df2.groupby(['factor']).describe()
```

```python
results = smf.ols('sale ~ C(factor)', data=df2).fit()
# results.summary()
aov_table = sms.anova_lm(results, typ=2)
aov_table
#(a.)Since the p-value<0.05, reject H0. There is sufficient evidence to show that there are differences in grading standards between the four high schools.
_ = df2['sale'].hist(by=df2['factor'], bins=7)
#print('the shape of A looks strange, so we delve into it')
stat, p = stats.shapiro(df2['sale'][df2['factor'] == 'A'])
print('Statistics=%.4f, p=%.4f' % (stat, p))
print("Since p-value > 0.05 = a, do not reject H0. it is normally distributed")
fig = sm.qqplot(df2['sale'][df2['factor'] == 'A'], stats.norm, 
                fit=True, line='45')
plt.show()
print("Since p-value > 0.05 = a, do not reject H0. The variances of the four treatments are equal.")
stats.bartlett(df2['sale'][df2['factor'] == 'A'], 
               df2['sale'][df2['factor'] == 'B'], 
               df2['sale'][df2['factor'] == 'C'],
               df2['sale'][df2['factor'] == 'D'])
print("(c.) Yes, it seems that all the requirements are satisfied")
```

上面的shapiro(常態), barlett(變異數相等)都要用H0,H1<br>

```python
print("(a.) Fisher’s LSD method")
alpha = 0.05
t = stats.t.ppf(1-alpha/2, 27) #27=30-3
LSD = t*(700*(1/10+1/10))**(1/2) #三組的樣本數一樣，LSD相等 700是MSE 10是那組內的數量
print(f"LSD : {LSD}")
print("compare 1&2 : ")
print("H0: μ1 = μ2  and H1: μ1 not equal to μ2")
print("128.7-101.4 = 27.3, Since 27.3>LSD, reject H0, the mean of group1 is not equal to group2")
print("")
print("compare 1&3 : ")
print("H0: μ1 = μ3  and H1: μ1 not equal to μ3")
print("133.7-128.7 = 5, Since 5<LSD, accept H0, There is no significant difference between the mean of group1 and group3")
print("")
print("compare 2&3 : ")
print("H0: μ2 = μ3  and H1: μ2 not equal to μ3")
print("133.7-101.4 = 32.3, Since 32.3>LSD, reject H0, the mean of group2 is not equal to group3")
print("")
print("the mean of group2 has a distance from group1 & group3")

print("(b.) Bonferroni Adjustment")
c = (3*2/2) #總共3組c3取2
alpha = 0.05/3
t = stats.t.ppf(1-alpha/2, 27)
LSD = t*(700*(1/10+1/10))**(1/2) #三組的樣本數一樣，LSD相等
print(f"LSD : {LSD}")

print("(c.) Tukey’s multiple comparison method")
q = statsmodels.stats.libqsturng.qsturng(0.95, 3, 27)
omega = q*((700/10)**(0.5)) #700是MSE
print(f"OMEGA : {omega}")
```

```python
_ = plt.hist(results.resid, bins='auto')
print("The histogram looks like a bell-shaped distribution.")
fig = sm.qqplot(results.resid, stats.norm, fit=True, line='45')
print("Q-Q Plot does show that almost all data points are close to the 45 degree line.")
stat, p = stats.shapiro(results.resid)
print('Statistics=%.4f, p=%.4f' % (stat, p))
print("The null hyptoehsis of Shapiro-Wilk test is that the data is normally distributed.")
print("The testing result shows that the null hyphosis cannot be rejected under 0.05 significance level.")
print("As a result, we conclude that the normality assumption is satisfied.")
```

# 兩變因相互影響ANOVA
H0: Two factors do not interact to affect the mean responses.<br>
H1: Two factors do interact to affect the mean responses <br>
<br>
H0: The means of time amount of Factor of "Drug" are equal <br>
H1: At least two means differ

```python
df = pd.read_excel('Xr14-108.xlsx')
display(df.head())
df = df.rename(columns = {  "Drug 1": "grade_a",
                            "Drug 2": "grade_b", 
                            "Drug 3": "grade_c", 
                            "Drug 4": "grade_d"}) 
df["id"] = df.index
#display(df)
df2 = pd.wide_to_long(df, ["grade"], sep = "_", suffix = '\w+', i="id", j="factor").reset_index()
display(df2.head())
result_interact = smf.ols('grade ~ C(factor) + C(Schedule) + C(factor):C(Schedule)', data=df2).fit()
# results.summary()
aov_interact = sms.anova_lm(result_interact, typ=2)
aov_interact
```

```python
#補上h0h1
print("the histogram looks like bell shape distribution")
print("Q-Q plot is also approximately satisfied")
#_ = plt.hist(result_interact.resid, bins='auto')
#fig = sm.qqplot(result_interact.resid, stats.norm, 
#                fit=True, line='45')
```

```python
# Shapiro-Wilk normality test. 補上
stat, p = stats.shapiro(result_interact.resid)
print('Statistics=%.4f, p=%.4f' % (stat, p))
print("we don't reject the null hypothesis")
df2['fac_Sch'] = df2.factor + "_" + df2.Schedule
# equal variance assumption
df2['resid'] = result_interact.resid
display(df2)


all_level = df2['fac_Sch'].unique()
print("List of all levels: ", all_level)

data_list = []
for alevel in all_level:
    data_list.append(df2['resid'][df2['fac_Sch'] == alevel])
print("There are %d samples collected" % len(data_list))

# Conduct Bartlett's Test
print("don't reject H0") #也要補一些H0H1
stats.bartlett(*data_list)
```

# 各項pi在母體中的比率是否=預期
H0 : pi = ai forall types <br>
H1 : At least 1 pi & ai differs<br>
這邊需要給更明確定義

```python
df = pd.read_excel('Xr15-13.xlsx')
display(df.head())
alpha = 0.05
nobs = df.shape[0]
df_v = df.value_counts()
df_v = pd.DataFrame(df_v, columns=['A'])
df_v = df_v.sort_values('Car type')
df_use= df_v.to_numpy()
df_use = df_use.reshape(1,-1)
prob_e = np.array([0.15, 0.25, 0.40, 0.20]) #預期機率
freq_e = nobs * prob_e #預期機率下預期數目
freq_o = df_use
freq_o = [i for k in freq_o for i in k]
freq_o #實際觀察數目
```

```python
if np.sum(freq_e < 5) > 0:
    print("Rule of five is not met. ")
else:
    stat, pvalue = stats.chisquare(freq_o, freq_e)
    print("Chi-squared test: statistics = %0.4f, p-value = %0.4f" % (stat, pvalue))

    df = len(freq_o)-1
    crit_value = stats.chi2.ppf(1 - alpha, df)
    print("Critical value = %0.4f (defree of freedom = %d)" % (crit_value, df))
#然後再給個結論
```

# 有沒有follow某種分布
degrees of freedom = k – p – 1, k = number of categories and p = number of population parameters estimated<br>
H0 : the data are normally distributed<br>
H1 : it is not

```python
alpha = 0.1
p1 = stats.norm.cdf(-1)
p2 = stats.norm.cdf(0)-p1
p3 = stats.norm.cdf(1) - stats.norm.cdf(0)
p4 = 1-stats.norm.cdf(1)
e = np.array([p1, p2, p3, p4])*50
f = np.array([6, 27, 14, 3])
print(e,"all above 5")
sum_all = 0
for i in range(e.shape[0]):
    #print(e[i],f[i])
    sum_all += ((e[i]-f[i])**2)/e[i]
sum_all
crit_value = stats.chi2.ppf(1 - alpha, 4-2-1)
print("Since",sum_all,">",crit_value, "\n we reject H0 in favor of H1 under 0.1 significance level,\nthe data are not normal.")
```

Krustal-Wallis - Wilcoxon的延伸，可以比較多個集合的location 可以不常態分佈，可以nominal<br>
Friedman 可以不常態，樣本不獨立是blocked<br>

```python
#省略先檢查常態的步驟考試記得補
#H0: The locations of all 3 populations are the same
#H1: At least two population locations differ
def grank(data1):
    if type(data1)==np.ndarray:
        alldata=data1.copy()
        data1=data1.copy()
    else:
        alldata=data1.values.copy()
        data1=data1.values.copy()
    alldata.sort()
    tmpdf=pd.DataFrame({'value':alldata})
    tmpdf['rank']=tmpdf.index+1
    v2r=tmpdf.groupby('value').mean().reset_index()
    samp1=pd.DataFrame({'value':data1})
    samp1=pd.merge(samp1,v2r,how='left')
    return samp1['rank']

grank(np.array([2,1,2,2]))
```

```python
df_xr1975 = pd.read_excel('Xr19-75.xlsx')
df_xr1975_2=df_xr1975.drop(['Job'],axis=1).apply(grank,axis=1)
display(df_xr1975_2)
b,k=df_xr1975_2.shape
Tall=df_xr1975_2.sum()
Fr=12/b/k/(k+1)*np.sum(Tall**2)-3*b*(k+1)
pvalue=1-stats.chi2.cdf(Fr,k-1)

print("Fr =",Fr)
print("pvalue =",pvalue)
```

```python
#另類整理資料集方式
df_xr1979 = pd.read_excel('Xr19-79.xlsx')
print('Head of dataset:')
display(df_xr1979.head())
alldata=np.concatenate([df_xr1979['4:00pm-Midnight'].to_numpy(),df_xr1979['Midnight-8:00 am'].to_numpy(),df_xr1979['8:00am-4:00 pm'].to_numpy()])
alldata.sort()
tmpdf=pd.DataFrame(({'value':alldata}))
tmpdf['rank']=tmpdf.index+1
value2rank=tmpdf.groupby('value').mean().reset_index()
value2rank
```

```python
samp1=pd.DataFrame({'value':df_xr1979['4:00pm-Midnight'].to_numpy()})
samp1=pd.merge(samp1,value2rank)
T1=samp1['rank'].sum()

samp2=pd.DataFrame({'value':df_xr1979['Midnight-8:00 am'].to_numpy()})
samp2=pd.merge(samp2,value2rank)
T2=samp2['rank'].sum()

samp3=pd.DataFrame({'value':df_xr1979['8:00am-4:00 pm'].to_numpy()})
samp3=pd.merge(samp3,value2rank)
T3=samp3['rank'].sum()

print("T1 = %f" % T1)
print("T2 = %f" % T2)
print("T3 = %f" % T3)
n1=df_xr1979['4:00pm-Midnight'].to_numpy().shape[0]
n2=df_xr1979['Midnight-8:00 am'].to_numpy().shape[0]
n3=df_xr1979['8:00am-4:00 pm'].to_numpy().shape[0]
n=n1+n2+n3
k=3
H=12/n/(n+1)*(T1**2/n1+T2**2/n2+T3**2/n3)-3*(n+1)
pvalue=1-stats.chi2.cdf(H,k-1)

print("H =",H)
print("pvalue =",pvalue)
```
