### (a.) bar chart   /      (b.)pie chart

```python
df_31 = pd.read_excel('Xr02-31.xlsx')
#display(df_31)
count_tab = df_31['Computer'].value_counts()
#display(count_tab)
CName = ("Dell", "Other", "HP", "Lenovo")
explode = (0.1, 0.1, 0.1, 0.1)

#plt.figure(figsize=(12, 8))
plt.pie(count_tab, explode=explode, labels=CName, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart for Computer Brands')

fig3, ax3 = plt.subplots()
rectsb = ax3.bar(CName, count_tab, width=0.5, bottom=None, align='center')
plt.ylabel('Frequency')
plt.title('Bar Chart for Computer Brands')
plt.xticks(CName)
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(rectsb)
plt.show()
```
