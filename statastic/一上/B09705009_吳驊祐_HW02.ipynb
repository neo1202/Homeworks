### 大部分老師能在10天以下的範圍送出

<div style="background-color:#D9D9D9;"><font size="5"> 4.9</font></div>

```python
import math as math
return_rate = pd.DataFrame([-0.15, -0.20, 0.15, -0.08, 0.50])
print("mean : ", return_rate.mean())
print("median : ",return_rate.median())
def geo_mean(rate):
    rate_1 = rate+1
    geo_m = math.exp(np.log(rate_1).mean()) - 1
    return geo_m
avg_rate = geo_mean(return_rate)
print("geo_mean = ", avg_rate)
```

## Geometric mean 是比較適合的，因為貼近實際上的平均金錢獲利比率
