```python
from matplotlib import pyplot as plt
%matplotlib inline
# 設定圖形大小; DPI越大圖越大
plt.rcParams["figure.dpi"] = 100
import seaborn as sns
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.stats.api as sms
import statsmodels.formula.api as smf
import statsmodels.stats.multicomp as smm
import statsmodels.stats.outliers_influence as sso
import statsmodels
import math
import statistics

def runsTest(l, l_median): 
    runs, n1, n2 = 1, 0, 0
    if(l[0]) >= l_median:
        n1 += 1   
    else:
        n2 += 1   
    for i in range(1, len(l)): 
        if (l[i] >= l_median and l[i-1] < l_median) or (l[i] < l_median and l[i-1] >= l_median):
            runs += 1  
        if(l[i]) >= l_median: 
            n1 += 1   
        else: 
            n2 += 1   
    runs_exp = ((2*n1*n2)/(n1+n2)) + 1
    stan_dev = math.sqrt((2*n1*n2*(2*n1*n2-n1-n2))/(((n1+n2)**2)*(n1+n2-1))) 
    z = (runs-runs_exp)/stan_dev
    pval_z = stats.norm.sf(abs(z)) * 2
    print('runs = ', runs)
    print('n1 = ', n1)
    print('n2 = ', n2)
    print('runs_exp = ', runs_exp)
    print('stan_dev = ', stan_dev)
    print('z = ', z)
    print('pval_z = ', pval_z)
    return pval_z

def Durbin_Watson_test(x):
    x_square_sum = np.vdot(x, x)
    print("x_square_sum = ", x_square_sum)
    size = x.size
    print("size = ", size)
    x_d = np.zeros((size))
    print("x_d = ", x_d)
    l_size = size - 1
    for i in range(l_size):
        x_d[i + 1] = x[i + 1] - x[i]
    print("x_d = ", x_d)
    d = np.vdot(x_d, x_d) / x_square_sum
    print("d = ", d)
    return(d)
```

```python
df = pd.read_excel('Xr17-01.xlsx')
display(df.tail())
y_var = df['Price'].values
xnames = ['Lot size', 'Trees', 'Distance']
df_a = df[['Price', 'Lot size', 'Trees', 'Distance']]
corr1 = df_a.corr() #沒差
y_data = df['Price']
X_data_ar = np.array([df['Lot size'], df['Trees'], df['Distance']])
X_data_T = X_data_ar.T
X_data = pd.DataFrame(X_data_T,columns = ['Lot size', 'Trees', 'Distance'])
X_data2 = sm.add_constant(X_data)
olsmod = sm.OLS(y_data, X_data2)
result_reg = olsmod.fit()
#print(result_reg.summary())

st, data, ss2 = sso.summary_table(result_reg, alpha=0.05)
#print("Columns in data are: %s" % ss2)
#Predicted value
y_pre = data[:, 2]
#Studentized Residual
SD = data[:, 10]
print("compute the residuals: ")
SD
```

To use the residual analysis, we need to apply below tests: <br>
#Normality Test <br>
#Heteroscedasticity and Homoscedasticity <br>
#Non-independence of the Error Variable<br>
<br>
Normality Test <br>
H0: Errors are normally distributed. <br>
H1: Errors are not normally distributed.

```python
# Test normality
print('Shapiro Test')
stat, p = stats.shapiro(SD)
print('Statistics=%.3f, p=%.3f' % (stat, p))
print("Because p_value = 0.141 > a =0.05, do not rejected H0. We can assume that errors are normally distributed.")
```

Heteroscedasticity and Homoscedasticity <br>
H0: When the requirement of a constant variance is not violated<br>
H1: When the requirement of a constant variance is violated

```python
plt.plot(y_pre, SD, 'o', color = 'gray')
plt.axhline(y=2, color = 'red', lw = 0.8)
plt.axhline(y=0, color = 'blue')
plt.axhline(y=-2, color = 'red', lw = 0.8)
plt.title('Standardized Residual Plot')
plt.xlabel('Predicted y value')
plt.ylabel('Standardized Residual')
plt.show()
print("Do not rejected H0. We can assume that the variation is constant and the mean is around 0.")
```

Non-independence of the Error Variable<br>
H0 : Randomness exists.<br>
H1 : Randomness does not exist.

```python
Id1 = data[:, 0]
plt.plot(Id1, SD, 'o', color = 'gray')
plt.axhline(y=0, color = 'blue')
plt.axhline(y=2, color = 'red')
plt.axhline(y=-2, color = 'red')
plt.title('Standardized Residual Plot')
plt.xlabel('Observation No.')
plt.ylabel('Standardized Residual')
plt.show()

SD_median = statistics.median(SD)
Z_pval = runsTest(SD, SD_median)   
print('p_value for Z-statistic= ', Z_pval)
print("Because p-value = 0.6024 > a = 0.05, do not reject H0 ")
print("There is no evidence to infer that the sample is not random.")
```
