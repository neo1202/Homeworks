```python
import gurobipy as gp 
from gurobipy import *
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
F = np.array([1200,1800,2500,1300])
P = np.array([1600,1000,1600,1200,1100,1400])

p1 = Model("p1")
x = p1.addVars(4,ub = 1, lb = 0, vtype = GRB.CONTINUOUS, name = 'x') #專利
y = p1.addVars(6,ub = 1, lb = 0, vtype = GRB.CONTINUOUS, name = 'y') #產品可否

p1.setObjective( quicksum(P[j]*y[j] for j in range(6))-quicksum(F[i]*x[i] 
                for i in range(4)), GRB.MAXIMIZE) 

p1.addConstr(y[0] <= x[0])
p1.addConstr(y[0] <= x[1])
p1.addConstr(y[0] <= x[3])
p1.addConstr(y[1] <= x[3])
p1.addConstr(y[2] <= x[1])
p1.addConstr(y[2] <= x[2])
p1.addConstr(y[3] <= x[0])
p1.addConstr(y[3] <= x[1])
p1.addConstr(y[3] <= x[3])
p1.addConstr(y[4] <= x[2])
p1.addConstr(y[4] <= x[3])
p1.addConstr(y[5] <= x[0])
p1.addConstr(y[5] <= x[1])
p1.addConstr((quicksum(F[i]*x[i] for i in range(4)) <= 4500 ), "預算")

p1.optimize()
for var in p1.getVars():
    print(var.varName, '= %.2f' %var.x)
print("\n")
print("objective value(max w) = %.2f" %p1.objVal)
```
