# Instance 2

## Problem1

```python
df = pd.read_excel("OR110-1_case01.xlsx","Instance 3")
display(df.head())

I = 4
J = df['Job ID'].shape[0]-1;
due = df['Due Time'].dropna();
split = df['Splitting Timing'].dropna();
M = 10000
N = 10000

for i in range(due.shape[0]):
    due[i+1] = due[i+1].hour + due[i+1].minute/60 - 7.5
#print(due)

processNum = [3,3,5,5,3,1,2,1,3,5] #Pj

processType = df[['Process Type','Unnamed: 2','Unnamed: 3','Unnamed: 4','Unnamed: 5']][1:] #Tj
processTime = df[['Processing Time','Unnamed: 8','Unnamed: 9','Unnamed: 10','Unnamed: 11']][1:] #Hjk

processType.rename(columns={'Unnamed: 5':'Process 5','Unnamed: 4':'Process 4','Unnamed: 3':'Process 3','Unnamed: 2':'Process 2','Process Type':'Process 1'},inplace = True)
processTime.rename(columns={'Unnamed: 11':'Process 5','Unnamed: 10':'Process 4','Unnamed: 9':'Process 3','Unnamed: 8':'Process 2','Processing Time':'Process 1'},inplace = True)
processTime = processTime.transpose()
processType = processType.transpose()
display(processType)
display(processTime)
```

```python
p1 = Model("p1")

c = []
z = []
a = []
y = []

w = p1.addVar(    lb = 0, vtype = GRB.CONTINUOUS, name = 'm')
c = p1.addVars(J, lb = 0, vtype = GRB.CONTINUOUS, name = 'c')    

z = p1.addVars(J,    lb = 0, vtype = GRB.BINARY, name = 'z')
a = p1.addVars(I, J, lb = 0, vtype = GRB.BINARY, name = 'a')
y = p1.addVars(J, J, lb = 0, vtype = GRB.BINARY, name = 'y') 
    
#print(*y, sep='\n')
```

```python
p1.setObjective(w , GRB.MINIMIZE) 

# add constraints and name them
p1.addConstrs((quicksum(a[i, j] for i in range(I)) == 1 for j in range(J)), "aij")    
p1.addConstr( (w    >= quicksum(z[j] for j in range(J))), "w >= zj")
p1.addConstrs((c[j] >= quicksum(processTime[j+1][k] for k in range(processNum[j])) for j in range(J)), "cj")    
p1.addConstrs((c[j] - due[j+1] <= N * z[j] for j in range(J)), "cj - Dj <= N * zj")
p1.addConstrs((due[j+1] - c[j] <= N *(1 - z[j]) for j in range(J)), "Dj - cj <= N * (1 - zj)")

p1.addConstrs((c[p] +  quicksum(processTime[q+1][k] for k in range(processNum[q])) - c[q] <= M * (    y[p, q] + 2 - a[i, p] - a[i, q]) for i in range(I) for p in range(J) for q in range(J) if p != q), "overlap q->p")
p1.addConstrs((c[q] +  quicksum(processTime[p+1][k] for k in range(processNum[p])) - c[p] <= M * (1 - y[p, q] + 2 - a[i, p] - a[i, q]) for i in range(I) for p in range(J) for q in range(J) if p != q), "overlap p->q")

p1.addConstr( (w    >= 0), "w >= 0" )
p1.addConstrs((c[j] >= 0 for j in range(J)), "j >= 0" )
```

### Stage 1

```python
p1.optimize()

# for var in p1.getVars():
#     print(var.varName, '= %.2f' %var.x)
print("\n")
print("objective value(min tardy jobs) = %.2f" %p1.objVal)
```

### Stage 2

```python
t = p1.objVal
m = p1.addVar(lb = 0, vtype = GRB.CONTINUOUS, name = 'm')


# set new objective function
p1.setObjective(m , GRB.MINIMIZE) 

# add more constraint
p1.addConstrs((m >= c[j] for j in range(J)), "m >= cj")
p1.addConstr((t >= w), "w <= t")
p1.addConstr((m >= 0 ), "m >= 0"  )

p1.optimize()

for var in p1.getVars():
    print(var.varName, '= %.2f' %var.x)
print("\n")
print("objective value(min makespan) = %.2f" %p1.objVal)
```

## Problem 2

```python
p2 = Model("p2")

J2 = 2*J #第一例為0-23, 0跟12為一組,11跟23為一組
L = np.zeros(J2) #Lj, 先分割每個job成兩份
display(split)
display(processNum)
for i in range(J): #0-11
    for k in range(0,int(split[i+1])):
        L[i] += processTime[i+1][k]
    for k in range(int(split[i+1]), processNum[i]):
        L[i+J] += processTime[i+1][k]
    #processTime第一格是1-12, k是0到processNum[j]
display(L[:J])
display(L[J:])
c = []
z = []
a = []
y = []

w = p2.addVar (    lb = 0, vtype = GRB.CONTINUOUS, name = 'w')
c = p2.addVars(J2, lb = 0, vtype = GRB.CONTINUOUS, name = 'c') #此處更動兩倍c    

z = p2.addVars(J,    lb = 0, vtype = GRB.BINARY, name = 'z')
a = p2.addVars(I, J2, lb = 0, vtype = GRB.BINARY, name = 'a') #有變
y = p2.addVars(J2, J2, lb = 0, vtype = GRB.BINARY, name = 'y') #0-23*0-23
```

```python
p2.setObjective(w , GRB.MINIMIZE) 

# add constraints and name them
p2.addConstrs((quicksum(a[i, j] for i in range(I)) == 1 for j in range(J2)), "aij")    
p2.addConstr( (w    >= quicksum(z[j] for j in range(J))), "w >= zj")
p2.addConstrs((c[j] >=  L[j] for j in range(J)), "cj")   #前半
p2.addConstrs((c[j+J] >=  L[j+J]+c[j] for j in range(J)), "cj+J")   #後半
p2.addConstrs((c[j+J] - due[j+1] <= N * z[j] for j in range(J)), "cj - Dj <= N * zj") #後半的cj
p2.addConstrs((due[j+1] - c[j+J] <= N *(1 - z[j]) for j in range(J)), "Dj - cj <= N * (1 - zj)") #後半的cj

p2.addConstrs((c[p] +  L[q] - c[q] <= M * (    y[p, q] + 2 - a[i, p] - a[i, q]) for i in range(I) for p in range(J2) for q in range(J2) if p != q), "overlap q->p")
p2.addConstrs((c[q] +  L[p] - c[p] <= M * (1 - y[p, q] + 2 - a[i, p] - a[i, q]) for i in range(I) for p in range(J2) for q in range(J2) if p != q), "overlap p->q")

p2.addConstr( (w    >= 0), "w >= 0" )
p2.addConstrs((c[j] >= 0 for j in range(J2)), "j >= 0" )
```

### Stage 1

```python
p2.optimize()

# for var in p1.getVars():
#     print(var.varName, '= %.2f' %var.x)
print("\n")
print("objective value(min tardy jobs) = %.2f" %p2.objVal)
```

### Stage 2

```python
t = p2.objVal
m = p2.addVar(lb = 0, vtype = GRB.CONTINUOUS, name = 'm')


# set new objective function
p2.setObjective(m , GRB.MINIMIZE) 

# add more constraint
p2.addConstrs((m >= c[j] for j in range(J2)), "m >= cj")
p2.addConstr((t >= w), "w <= t")
p2.addConstr((m >= 0 ), "m >= 0"  )

p2.optimize()

for var in p2.getVars():
    print(var.varName, '= %.2f' %var.x)
print("\n")
print("objective value(min makespan) = %.2f" %p2.objVal)
```
