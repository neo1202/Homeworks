```python
# -*- coding: utf-8 -*-
"""
OR Final Project: Group A
"""
# Q1(b)
from gurobipy import*
import pandas as pd

model = Model('OR Final Project')

# In[] Sets

# 站點集合
#I = ["公館捷運站", "社科院", "長興街", "總圖"]
I = 4

# In[] Parameters
# 順序爲 ： ["公館捷運站", "社科院", "長興街", "總圖"]

# 站點 i 的停車數量
Parked = [100,65,110,10]

# 為站點 i 的車柱總數
Capacity = [200, 100, 150, 20]

# 調度車輛從站點 i 到站點 j 的成本
# c = [[1, 6, 5, 3], [6, 1, 4, 1], [3, 1, 1, 2], [5, 2, 4, 1]]
c = [[1 for i in range(I)] for j in range(I)]

# 一台調度車容量上限
A = 16

# 停車率的安全範圍上限比率
S_H = 0.7

# 停車率的安全範圍下限比率
S_L = 0.3

# 調度車在單位時間內移動的距離（或時間）上限
L = 60

# λ 為各站沒補足下限或補超過上限的懲罰係數
lambda_value = 0.5

# In[] Assign variables
a=[]
b=[]
x=[]
y=[]
z=[]
w_L=[]
w_H=[]

a = model.addVars(I,    lb = 0, vtype = GRB.INTEGER, name = 'a')
b = model.addVars(I,    lb = 0, vtype = GRB.INTEGER, name = 'b')
z = model.addVars(I,    lb = 0, vtype = GRB.INTEGER, name = 'z')
w_L = model.addVars(I,  lb = 0, vtype = GRB.INTEGER, name = 'w_L')
w_H = model.addVars(I,  lb = 0, vtype = GRB.INTEGER, name = 'w_H')

x = model.addVars(I,    lb = 0, vtype = GRB.BINARY, name = 'x')
y = model.addVars(I, I, lb = 0, vtype = GRB.BINARY, name = 'y') 


model.update()

# In[] Assigning objective and constraints

model.setObjective( quicksum(y[i,j]*c[i][j] for i in range(I) for j in range(I)) + lambda_value*quicksum(w_L[i]+w_H[i] for i in range(I)) , GRB.MINIMIZE)

# C1
model.addConstrs((x[i] + x[j]  >= 2*y[i,j] for i in range(I) for j in range(I)), "C1")

# C1'
model.addConstrs((quicksum(y[i,j] for j in range(I)) +  quicksum(y[j,i] for j in range(I))>= x[i] for i in range(I) ), "C1-1'")
# model.addConstrs((y[i,j] > x[i] for i in range(I) for j in range(I)), "C1-2'")

# C2
model.addConstr((quicksum(a[i] for i in range(I)) >= quicksum(z[i] for i in range(I))), "C2-1")

# C2'
model.addConstrs((a[i]*x[i] - b[i]*x[i] + z[i] <= A for i in range(I)), "C2-1'")
model.addConstrs((a[i]*x[i] - b[i]*x[i] + z[i] >= 0 for i in range(I)), "C2-2'")

# C3'
model.addConstrs( (y[i,j] * (z[i]+a[i]-b[i]) <= y[i,j] * z[j] for i in range(I) for j in range(I) ) , "C3'")

# C3
# model.addConstr((z[0] == 0), "C3") 
    
# C4
M_a = sum(Parked)
M_b = sum(Parked)
M_z = sum(Parked)
model.addConstrs((a[i] <= M_a * x[i] for i in range(I)), "C4-1")
model.addConstrs((b[i] <= M_b * x[i] for i in range(I)), "C4-2")
model.addConstrs((z[i] <= M_b * x[i] for i in range(I)), "C4-3")


# C5
model.addConstrs((w_L[i] >= S_L - (Parked[i] - a[i] * x[i] + b[i] * x[i])/Capacity[i] for i in range(I)), "C5-1")
model.addConstrs(((Parked[i] - a[i] * x[i] + b[i] * x[i])/Capacity[i] >= S_L for i in range(I)), "C5-2")

# C6
model.addConstrs((w_H[i] >= (Parked[i] - a[i] * x[i] + b[i] * x[i])/Capacity[i] - S_H for i in range(I)), "C6-1")
model.addConstrs(((Parked[i] - a[i] * x[i] + b[i] * x[i])/Capacity[i] <= S_H for i in range(I)), "C6-2")

# C7
model.addConstr( (quicksum(y[i,j]*c[i][j] for i in range(I) for j in range(I)) <= L) , "C7")

# C8
model.addConstr((quicksum(x[i]*a[i] for i in range(I)) == quicksum(x[i]*b[i] for i in range(I))), "C8")
                
# C9
model.addConstrs((b[i] * x[i] + Parked[i] <= Capacity[i] for i in range(I)), "C9")

# C10
model.addConstrs((a[i] * x[i] <= Parked[i] for i in range(I)), "C10")

# C11
model.addConstrs((y[i,j] + y[j,i] <= 1 for i in range(I) for j in range(I)), "C11")

# C12
model.addConstrs((y[i,j] == 0 for i in range(I) for j in range(I) if i == j), "C12")

# trial constr
model.addConstrs( ( quicksum(y[i,j] for i in range(I)) <= 1 for j in range(I) ) , "trial")
model.addConstrs( ( quicksum(y[i,j] for j in range(I)) <= 1for i in range(I) ) , "trial")

model.optimize()

# In[] Export result
print('Objective value: %2f'%(model.objVal))

output = pd.DataFrame(columns = ("Var", "Value"))
no = 1
for _,v in enumerate(model.getVars()):
    output.loc[no] = [v.varname, float("{:.4f}".format(v.x))]
    no += 1
    
output = pd.DataFrame(output)
print(output)

```
